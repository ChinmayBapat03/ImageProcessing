```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage # Multi-dimensional image processing
img=mpimg.imread('/content/me.png')
blurred_face=ndimage.gaussian_filter(img, sigma=1.5) # Multidimensional Gaussian filter
plt.figure(1)
plt.imshow(img,cmap='gray')
plt.figure(2)
plt.imshow(blurred_face,cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage # Multi-dimensional image processing
img=mpimg.imread('/content/me.png')
blurred_face=ndimage.laplace(img) # N-dimensional Laplace filter based on approximate second derivatives
plt.figure(1)
plt.imshow(img,cmap='gray')
plt.figure(2)
plt.imshow(blurred_face,cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import ndimage # Multi-dimensional image processing
img=mpimg.imread('/content/me.png')
blurred_face=ndimage.median_filter(img,5) # Calculate a multidimensional median filter
plt.figure(1)
plt.imshow(img,cmap='gray')
plt.figure(2)
plt.imshow(blurred_face,cmap='gray')
```
